```python
plt.imshow(arr)
```

# 1. Chess Board

```python
zero = np.zeros((8,8), dtype= int)

for i in range(len(zero)):
    for j in range(len(zero[i])):
        if (i+j)%2 == 0:
            zero[i][j] = 255

zero
```

```python
plt.imshow(zero)
```

```python
cv.imwrite("Chess.png", zero)
```

# 2. High pixel density image

```python
img = []
for i in range(720):
    temp = []
    for j in range(1280):
        temp.append(255)
    
    img.append(temp)

img = np.array(img)
plt.imshow(img)
```
