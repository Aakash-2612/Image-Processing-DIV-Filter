# Creating Dense Shades of RGB

```python
img = []
for i in range(256):
    t = []
    for j in range(256):
        t.append([i,0,0])
    img.append(t)
    
img = np.array(img)
plt.imshow(img)
```

```python
img = []
for i in range(256):
    t = []
    for j in range(256):
        t.append([0,i,0])
    img.append(t)
    
img = np.array(img)
plt.imshow(img)
```

```python
img = []
for i in range(256):
    t = []
    for j in range(256):
        t.append([0,0,i])
    img.append(t)
    
img = np.array(img)
plt.imshow(img)
```

# Merging the Dense Shades

```python
img = []

# Red

for i in range(256):             # From Black to Red
    t = []
    for j in range(1536):
        t.append([i,0,0])
    img.append(t)

for i in range(255, 0, -1):      # From Red to Black
    t = []
    for j in range(1536):
        t.append([i,0,0])
    img.append(t)
    
# Green

for i in range(256):             # From Black to Green
    t = []
    for j in range(1536):
        t.append([0,i,0])
    img.append(t)

for i in range(255, 0, -1):      # From Green to Black
    t = []
    for j in range(1536):
        t.append([0,i,0])
    img.append(t)

# Blue

for i in range(256):             # From Black to Blue
    t = []
    for j in range(1536):
        t.append([0,0,i])
    img.append(t)

for i in range(255, 0, -1):      # From Blue to Black
    t = []
    for j in range(1536):
        t.append([0,0,i])
    img.append(t)
    
img = np.array(img)
plt.imshow(img)
cv.imwrite('Transition.jpg', img)
```
